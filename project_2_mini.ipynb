```python
import pandas as pd 

df = pd.read_excel(r'C:\Users\SYAMNARAYANAN\OneDrive\Desktop\visual code\vaccine-introduction-data.xlsx')


df
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.dropna(subset=["countryname", "year"], inplace=True)
df[["description","who_region"]].fillna("Unknown", inplace=True)
```

```python
df["year"] = df["year"].astype(int)
```

```python
df = df[df["iso_3_code"].str.len() == 3]
```

```python
df["ISO_3_CODE"] = df["ISO_3_CODE"].str.upper().str.strip()
df["COUNTRYNAME"] = df["COUNTRYNAME"].str.title().str.strip()
df["WHO_REGION"] = df["WHO_REGION"].str.upper().str.strip()
df["DESCRIPTION"] = df["DESCRIPTION"].str.strip()
```

```python
df['intro'].value_counts(dropna=False)
```

```python
intro_mapping = {
    "Yes": True,
    "Yes (R)": True,
    "Yes (P)": True,
    "Yes (A)": True,
    "Yes (O)": True,
    "Yes (D)": True,
    "No": False,
    "ND": None,
    "High risk area": None
}

df['intro'] = df['intro'].map(intro_mapping)
df['intro'].value_counts(dropna=False)
```

```python
df.to_csv("cleaned_vaccination_data.csv", index=False)
```

```python
import sqlite3

conn = sqlite3.connect("vaccination_data.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS VaccinationData (
    iso_3_code TEXT NOT NULL,
    country_name TEXT NOT NULL,
    who_region TEXT NOT NULL,
    year INTEGER NOT NULL,
    description TEXT NOT NULL,
    intro BOOLEAN
)
""")
conn.commit()
print("Table created (if not already exists)")
```

```python
df.to_sql("VaccinationData", conn, if_exists="replace", index=False)
print("Cleaned data successfully saved to SQLite database")
```

```python
df5 = pd.read_excel(r"C:\Users\SYAMNARAYANAN\OneDrive\Desktop\visual code\coverage-data (1).xlsx")

df5
```

```python
df5.columns = df5.columns.str.strip().str.lower()
```

```python
df5.isnull().sum()  
```

```python
df5.rename(columns={'code': 'iso_3_code','group':'group_name','name':'country_name'}, inplace=True)
```

```python
cat_cols = [
    'iso_3_code', 'country_name', 'antigen', 'antigen_description',
    'coverage_category', 'coverage_category_description'
]
df5[cat_cols] = df5[cat_cols].fillna('Unknown')

num_cols = ['target_number', 'doses', 'coverage']
df5[num_cols] = df5[num_cols].fillna(0)

df5 = df5.dropna(subset=['year'])

df5['year'] = df5['year'].astype(int)
```

```python
df5['target_number'] = df5['target_number'].round(0).astype('int')
df5['doses'] = df5['doses'].round(0).astype('int')
df5['coverage'] = df5['coverage'].round(0).astype('int')
```

```python
df5['doses'] = df5['doses'].astype(int)

```

```python
df5.info()
```

```python
df5.isnull().sum()   
```
